```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
collision_data = pd.read_csv('collisions.csv', parse_dates=['timestamp'])

def visualize_collisions(time_range='H'):
    grouped_data = collision_data.set_index('timestamp').groupby(pd.Grouper(freq=time_range)).size()

    plt.figure(figsize=(10, 6))
    grouped_data.plot(kind='bar')
    plt.title(f'Collisions Grouped by {time_range} ~ same penalty to the waiting vehicle')
    plt.xlabel('Time')
    plt.ylabel('Number of Collisions')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

visualize_collisions('5S')
```

```python
# Read the data from the file
waitimg_duration_data = pd.read_csv('max_wait_durations.csv')

def visualize_waiting_durations():
    # Convert timestamp to datetime
    waitimg_duration_data['timestamp'] = pd.to_datetime(waitimg_duration_data['timestamp'])

    # Plotting the data
    plt.figure(figsize=(12, 6))
    plt.plot(waitimg_duration_data['timestamp'], waitimg_duration_data['max_waiting_duration'], marker='o', linestyle='-')
    plt.xlabel('Timestamp')
    plt.ylabel('Max Waiting Duration')
    plt.title('Max Waiting Duration Over Time With Same Waiting Penalty')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()

    # Show plot
    plt.show()

visualize_waiting_durations()
```
